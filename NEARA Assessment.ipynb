# FOLIUM Iteration

```python
import folium
import json
import requests

# Define data from json
conductors = 'Conductors.json'
conductors = json.load(open(conductors))
poles = 'Poles.json'
poles = json.load(open(poles))



# Load map at location
m = folium.Map(location=[33.92544, -116.73645], zoom_start=14)

folium.map.FeatureGroup(name='Poles').add_to(m)


folium.map.FeatureGroup(name='Conductors').add_to(m)




# Load Conductor Data
folium.GeoJson(conductors,
               style_function=lambda feature: {
                              'fillColor': 'green',
                              'color': 'darkred',
                              'weight': 0.75,
                              'name': 'Conductors'
               },
               tooltip=folium.features.GeoJsonTooltip(
                               fields=['COND_ID']
               )
              ).add_to(m)




# Load Pole Data
folium.GeoJson(poles,
               style_function=lambda feature: {
                              'fillColor': 'green',
                              'color': 'black',
                              'weight': 10,
                              'name': 'poles'
               },
               tooltip=folium.features.GeoJsonTooltip(
                               fields=['POLE_NAME_']
               )
               ).add_to(m)



folium.LayerControl().add_to(m)


# FeatureGroup(name='Conductors')


m



# features_df = pd.DataFrame(data['features'])
# x = features_df['properties']
# y = features_df['geometry']

# features_df
# x = pd.DataFrame(x['features'])
# x
```

# MAPBOX Iteration

```python
import numpy as np
import json
import plotly.graph_objects as go
import pandas as pd
from operator import itemgetter

mapbox_access_token = open("mapbox_token.txt").read()

# Opening file
r = open('Poles.json')
# Opening file
s = open('Conductors.json')

# returns JSON object as a dictionary
poles = json.load(r)
poles

cond = json.load(s)
cond

# Normalize Poles Data
pole_df = pd.json_normalize(poles, record_path = ['features'])
pole_df

# Normalize Conduit Data
cond_df = pd.json_normalize(cond, record_path = ['features'])
cond_df

pole_coords = pole_df.loc[:, 'geometry.coordinates']
pole_coords

cond_coords = cond_df.loc[:, 'geometry.coordinates']
cond_coords


# https://www.geeksforgeeks.org/python-get-first-element-of-each-sublist/
def pole_lon(pole_coords):
    return [item[0] for item in pole_coords]
def pole_lat(pole_coords):
    return [item[1] for item in pole_coords]


# https://stackoverflow.com/questions/72256826/iterate-through-list-in-nested-list-python
flatlist = [el for lon in cond_coords for lat in lon for el in lat]
cond_lon, cond_lat = flatlist[0::2], flatlist[1::2]



fig1 = go.Figure()
# https://plotly.com/python/lines-on-mapbox/
fig1.add_trace(go.Scattermapbox(
        lat = cond_lat,
        lon = cond_lon,
        mode = 'lines',
        marker = go.scattermapbox.Marker(
            size = 20,
            color = 'Black',
            opacity = 0.25
        ),
        text = cond_df['properties.COND_ID'],
#         hovertemplate = '<b>Conduit ID #:cond_df<br>'
    ))

fig1.add_trace(go.Scattermapbox(
        lat = pole_lat(pole_coords),
        lon = pole_lon(pole_coords),
        mode = 'markers',
        marker = go.scattermapbox.Marker(
            size = 20,
            color = '#0000ff',
            opacity = 0.25
        ),
        text = pole_df['properties.POLE_NAME_'],
#         hovertemplate = '<b>Pole Serial #:</b>: null<br>'

    ))


fig1.update_layout(
    autosize=True,
    hovermode='closest',
    mapbox=dict(
        accesstoken=mapbox_access_token,
        bearing=0,
        center=dict(
            lat=33.92544,
            lon=-116.73665
        ),
        style="stamen-terrain",
        pitch=0,
        zoom=13.5
    ),
)
fig1
```
